```python
import numpy as np
```

```python
x_train = np.array([1.0, 2.0])           #(size in 1000 square feet)
y_train = np.array([300.0, 500.0])           #(price in 1000s of dollars)
```

```python
#Now let's consider how to do this
def compute_model_output(x, w, b):
    '''
    Here's an example computation 

    x = [1.0, 2.0]
    w = 2.0
    b = 1.0 

    y = 2.0* [1.0, 2.0] + 1.0
    y = [2.0, 4.0] + 1.0
    y = [3.0, 5.0]

    '''

    # shape is (# rows, # cols)
    m = x.shape[0]
    out = np.zeros((m))
    for i in range(m):
        out[i] = w*x[i] + b
    return out

def cost(x, w, b, y):
    '''
    x: m X 1 vector 
    w: weight 
    b: scalar 
    y: m + x bector 
    '''
    y_pred = compute_model_output(x,w,b)
    # of rows 
    m = y_pred.shape[0]
    cost = 0 
    for i in range(m):
        diff = (y_pred[i] - y[i])**2
        cost += diff 
    #Now at the end we need to combine the two
    cost *= (1/(2*m))
    return cost
```

```python
def compute_gradient(x,w,b,y):
    '''
    ''' 
    #use the formula  dJ/db = 1/m  (sum  (y_{hat}_{i} - y_i))
    db = 0 
    dw = 0 
    m = x.shape[0]
    for i in range(m):
        y_pred_i = w*x[i] + b 
        y_diff =  y_pred_i - y[i]
        dw += y_diff*x[i]
        db += y_diff 
    db *= (1/ m)
    dw *= (1/ m)
    return (dw, db)
    # use the formula dJ/dw = 1/m (sum(y_{hat}_{i} - y_{i})(x_{i}))

def gradient_descent(x: np.array, y: np.array,w,b, iters: int, learning_rate: float):
    x_init = x 
    y_init = y 
    for i in range(iters):
        print(f"current loss: {str(cost(x,w,b, y))}")
        dw, db = compute_gradient(x,w,b,y)
        w = w - learning_rate*dw 
        b = b - learning_rate*db
    return w,b
```
